```python
import pandas as pd
import sqlite3 as sql
#Merged parts of tweets about Berkshire hathaway before pandemic

BH_tweets_before_pandemic_part1 = pd.read_csv('C:/Users/ll/Desktop/dane/BH_tweets_before_pandemic_part1.csv', sep = '\t')
BH_tweets_before_pandemic_part2 = pd.read_csv('C:/Users/ll/Desktop/dane/BH_tweets_before_pandemic_part2.csv', sep = '\t')
BH_tweets_before_pandemic_part3 = pd.read_csv('C:/Users/ll/Desktop/dane/BH_tweets_before_pandemic_part3.csv', sep = '\t')
BH_tweets_before_pandemic_part4 = pd.read_csv('C:/Users/ll/Desktop/dane/BH_tweets_before_pandemic_part4.csv', sep = '\t')





con = sql.connect('example.db')
BH_tweets_before_pandemic_part1.to_sql('BH_tweets_before_pandemic_part1', con, if_exists='replace')
BH_tweets_before_pandemic_part2.to_sql('BH_tweets_before_pandemic_part2', con, if_exists='replace')
BH_tweets_before_pandemic_part3.to_sql('BH_tweets_before_pandemic_part3', con, if_exists='replace')
BH_tweets_before_pandemic_part4.to_sql('BH_tweets_before_pandemic_part4', con, if_exists='replace')

BH_tweets_before_pandemic = pd.read_sql("SELECT * FROM BH_tweets_before_pandemic_part1 UNION SELECT * FROM BH_tweets_before_pandemic_part2 UNION SELECT * FROM BH_tweets_before_pandemic_part3 UNION SELECT * FROM BH_tweets_before_pandemic_part4;", con)


BH_tweets_before_pandemic.to_sql('BH_tweets_before_pandemic', con, if_exists='replace', index = False)

BH_tweets_before_pandemic
```
