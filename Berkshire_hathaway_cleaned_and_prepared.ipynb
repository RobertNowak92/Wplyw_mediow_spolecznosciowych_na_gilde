```python
import pandas as pd
import numpy as np
from textblob import TextBlob
from textblob.sentiments import PatternAnalyze
```

```python
Berkshire_hathaway_tweets = pd.read_csv('dane/Berkshire_hathaway_tweets.csv')
```

```python
#dodanie formatu daty
Berkshire_hathaway_tweets['created_at'] = Berkshire_hathaway_tweets['created_at'].str.replace(' CET', '')
Berkshire_hathaway_tweets['created_at'] = pd.to_datetime(Berkshire_hathaway_tweets['created_at'], format = '%Y-%m-%d %H:%M:%S')
```

```python
#wybór tweetów tylko w języku angielskim
Berkshire_hathaway_tweets_clean = Berkshire_hathaway_tweets[Berkshire_hathaway_tweets.language == 'en']
Berkshire_hathaway_tweets_clean = Berkshire_hathaway_tweets_clean.reset_index()
Berkshire_hathaway_tweets_clean.shape
```

```python
#usunięcie column które nie mają żadnych wartości
Berkshire_hathaway_tweets_clean = Berkshire_hathaway_tweets_clean.dropna(thresh = 1, axis = 1)
```

```python
Berkshire_hathaway_tweets_clean.info()
```

```python
Berkshire_hathaway_tweets_clean.head()
```

```python
#sprawdzenie wartości atrybutu place, zmienna jest najprawdopodobniej nieistotna
Berkshire_hathaway_tweets_clean['place'].value_counts()
```

```python
#sprawdzenie wartości atrybutu quote_url
Berkshire_hathaway_tweets_clean['quote_url'].value_counts()
```

```python
#podliczenie ilości linków i zapisanie ich
Berkshire_hathaway_tweets_clean['quote_url'] = Berkshire_hathaway_tweets_clean['quote_url'].str.count('https://')

#zastąpienie NaN przez 0
Berkshire_hathaway_tweets_clean.fillna({'quote_url':0}, inplace=True)

#zmiana typu danych z float64 na int64
Berkshire_hathaway_tweets_clean['quote_url'] = Berkshire_hathaway_tweets_clean['quote_url'].astype(np.int64)
```

```python
#zmiana NaN na [] w columnie thumbnail
Berkshire_hathaway_tweets_clean.fillna({'thumbnail':'[]'}, inplace=True)
```

```python
#brak NaN poza columną place która zostanie później usunięta
Berkshire_hathaway_tweets_clean.info()
```

```python
#szukanie duplikatów
Berkshire_hathaway_tweets_clean[Berkshire_hathaway_tweets_clean.duplicated(subset = 'id')]
```

```python
#atrybut retweet nie posiada żadnych wartości przydatnych do analizy
Berkshire_hathaway_tweets_clean['retweet'].value_counts()
```

```python
#usunięcie niepotrzebnych column
Berkshire_hathaway_tweets_clean = Berkshire_hathaway_tweets_clean.drop(['conversation_id', 
                                                                        'date', 'time', 'timezone', 
                                                                        'language', 'retweet', 'link','place'], 
                                                                       axis = 1)
```

```python
#podliczenie ilości elementów dla predyktorów przechowywanych jako lista
columns_to_format = ['mentions', 'urls', 'photos', 'hashtags', 'cashtags', 'thumbnail']
for column in columns_to_format:
    empty_list = Berkshire_hathaway_tweets_clean[column] == '[]'
    Berkshire_hathaway_tweets_clean[column] = Berkshire_hathaway_tweets_clean[column].str[1:-1].str.split(',')
    Berkshire_hathaway_tweets_clean[column] = Berkshire_hathaway_tweets_clean[column].str.len()
    Berkshire_hathaway_tweets_clean.loc[empty_list, column] = 0
```

```python
#podliczenie ilości elementów słownika w reply_to i zapisanie ich
Berkshire_hathaway_tweets_clean['reply_to'] = Berkshire_hathaway_tweets_clean['reply_to'].str.count('{')

#zmiana typu danych z float64 na int64
Berkshire_hathaway_tweets_clean['reply_to'] = Berkshire_hathaway_tweets_clean['reply_to'].astype(np.int64)
```

```python
Berkshire_hathaway_tweets_clean['reply_to'].value_counts()
```

```python
Berkshire_hathaway_tweets_clean.info()
```

```python
#określenie ilości słów w tweetach
Berkshire_hathaway_tweets_clean['tweet_word_count'] = Berkshire_hathaway_tweets_clean['tweet'].str.split().apply(len)
```

```python
#Analiza opini na podstawie pakietu textblob
for i in range(Berkshire_hathaway_tweets_clean.shape[0]):
    blob = TextBlob(Berkshire_hathaway_tweets_clean.loc[i, 'tweet'], analyzer=PatternAnalyzer())
    Berkshire_hathaway_tweets_clean.loc[i, 'polarity'] = blob.sentiment[0]
    Berkshire_hathaway_tweets_clean.loc[i, 'subjectivity'] = blob.sentiment[1]
```

```python
Berkshire_hathaway_tweets_clean.info()
```

```python
#zapis wyczyszczonej i pr
Berkshire_hathaway_tweets_clean.to_csv('dane/Berkshire_hathaway_tweets_clean.csv')
```
