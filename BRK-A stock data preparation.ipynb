```python
import pandas as pd
```

```python
#Check stock data for Berkshire hathaway
BRK_A = pd.read_csv('dane/BRK-A_2017-09-29_to_2022-04-01.csv')
BRK_A
```

```python
BRK_A.describe()
```

```python
#Added date format and changed index to a date
BRK_A['Date'] =  pd.to_datetime(BRK_A['Date'] , format='%Y-%m-%d')
BRK_A.set_index('Date', inplace=True)
```

```python
BRK_A.info()
```

```python
#Added missing dates
idx = pd.date_range('2017-09-29', '2022-04-01')
BRK_A = BRK_A.reindex(idx)
BRK_A.index.name = 'date'
```

```python
BRK_A
```

```python
#filled values for missing dates by liner function between previous 
#known value and last known value for each missing periods
column_names = BRK_A.columns
for i in BRK_A.index:
    itemp = []
    if pd.isnull(BRK_A.loc[i ,'Open']):
        itemp.append(i)
        while pd.isnull(BRK_A.loc[itemp[-1] ,'Open']):
            itemp.append(itemp[-1] + pd.Timedelta(days=1))
        for col in column_names:
            diff = (BRK_A.loc[itemp[-1], col] - BRK_A.loc[itemp[0]-pd.Timedelta(days=1), col])/len(itemp)
            for date in itemp:
                BRK_A.loc[date, col] = BRK_A.loc[date-pd.Timedelta(days=1), col] + diff
```

```python
#Droped column Adj Close
BRK_A = BRK_A.drop(columns = ['Adj Close'])
```
